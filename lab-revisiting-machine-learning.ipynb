1. Creating a dataframe with the percentages of null values per feature

```python
nulls = round(data.isnull().sum()*100/len(data), 2)
nulls = pd.DataFrame(nulls, columns= ['null_percentage'])
nulls.sort_values(by = 'null_percentage', ascending=False)
```

3. Removing columns with more than 85% NaNs

```python
# retrieving the columns
nulls_85 = nulls[nulls['null_percentage'] > 85]
```

```python
# turning nulls_85 into a list containing only the names of the columns to use the new list in a .drop() function
nulls_85_cols = list(nulls_85.T.columns)
nulls_85_cols
```

```python
# dropping the columns
data = data.drop(nulls_85_cols, axis = 1)
data
```

3. Cleaning GENDER column

```python
# checking out the distinct values for GENDER
data.GENDER.unique()
```

```python
# creating a function that will change every value != "F" or "M" to "other" in column GENDER
def gender_clean(x):
    if x == "F" or x == "M":
        return x
    else:
        return "other"
```

```python
# applying function and checking to see if it worked
data['GENDER'] = data['GENDER'].apply(lambda x : gender_clean(x))
data['GENDER'].unique()
```
